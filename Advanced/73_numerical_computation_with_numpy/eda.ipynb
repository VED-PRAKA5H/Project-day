## Broadcasting and Scalars
Now, oftentimes you'll want to do some sort of operation between an array and a single number. In mathematics, this single number is often called a scalar.
> In order to achieve this result, NumPy will make the shape of the smaller array - our scalar - compatible with the larger array. This is what the documentation refers to when it mentions the term `broadcasting`.

The same rules about 'expanding' the smaller ndarray hold true for 2 or more dimensions. We can see this with a 2-Dimensional Array:

````python
array_2d = np.array([[1, 2, 3, 4], 
                     [5, 6, 7, 8]])
````

````python
array_2d + 10
````

````python
array_2d * 10
````


Broadcasting also works with arrays of compatible shapes following these general rules:

- Dimensions are compared from the trailing (rightmost) dimension.
- Two dimensions are compatible if they are equal or one of them is 1.
- Arrays are stretched along the dimension with size 1.

Examples from the documentation:

| A shape         | B shape      | Result shape    |
|-----------------|--------------|-----------------|
| (5, 4)          | (1,)         | (5, 4)          |
| (5, 4)          | (4,)         | (5, 4)          |
| (15, 3, 5)      | (15, 1, 5)   | (15, 3, 5)      |
| (15, 3, 5)      | (3, 5)       | (15, 3, 5)      |
| (15, 3, 5)      | (3, 1)       | (15, 3, 5)      |

## Matrix Multiplication with @ and .matmul()

Matrix multiplication follows [linear algebra rules](https://en.wikipedia.org/wiki/Matrix_multiplication#Illustration). Here, the result matrix element $ c_{ij} $ is the dot product of row $ i $ from the first matrix and column $ j $ from the second matrix.

Let's define two matrices \( a1 \) and \( b1 \):

<img src=https://i.imgur.com/LCdhmi8.png width=350>

````python
a1 = np.array([[1, 3],
               [0, 1],
               [6, 2],
               [9, 7]])

b1 = np.array([[4, 1, 3],
               [5, 8, 5]])

print(f'{a1.shape}: a has {a1.shape[0]} rows and {a1.shape[1]} columns.')
print(f'{b1.shape}: b has {b1.shape[0]} rows and {b1.shape[1]} columns.')
print('Dimensions of result: (4x2)*(2x3)=(4x3)')
````

**Challenge**: Let's multiply `a1` with `b1`. Looking at the wikipedia example above, work out the values for c12 and c33 on paper. Then use the [`.matmul()`](https://numpy.org/doc/stable/reference/generated/numpy.matmul.html) function or the `@` operator to check your work. 

````python
# Matrix multiplication using matmul
c1 = np.matmul(a1, b1)
c1
````

````python
# Alternatively, using the `@` operator:
c2 = a1 @ b1
c2
````

````python
# Verify, Both methods should result in the same matrix
assert np.array_equal(c1, c2)
````

## Challenge 2: Converting to Grayscale
Convert the image to black and white. The values in our `img` range from 0 to 255. \
To convert an RGB image to grayscale, you can use the **perceptual luminance-preserving** formula:

$
Y_{linear} = 0.2126 \times R_{linear} + 0.7152 \times G_{linear} + 0.0722 \times B_{linear}
$

$Y_{linear}$ is what we're after - our black and white image.\
However, this formula only works if our red, green and blue values are between 0 and 1 - namely in sRGB format.\
Currently the values in our img range from 0 to 255. So:
* Divide all the values by 255 to convert/normalize them to sRGB, where all the values are between 0 and 1. 
* Next, multiply the sRGB array by the `grey_vals` to convert the image to grey scale. 
* Finally use Matplotlib's [`.imshow()`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.imshow.html) together with the colormap parameter set to gray `cmap=gray` to look at the results. 

````python
# Normalize pixel values to sRGB between 0 and 1
sRGB_array = img / 255.0
````

````python
# Coefficients for RGB to grayscale conversion per perception standards
grey_vals = np.array([0.2126, 0.7152, 0.0722])
````

````python
# Convert to grayscale using matrix multiplication (dot product)
img_gray = sRGB_array @ grey_vals
````
